由于从tushare上获取的数据存在缺失值,需要对缺失值进行处理
下面记载在这过程中遇到的坑(以下记载的方法可以用于不同情况的插值处理)

首先导入相关模块

```python
import tushare as ts
import pandas as pd
```

设置抓取的股票时间区间和股票代码(symbol)

```python
start,end,stockcode = '2013-10-19','2016-10-19','000002'
```

通过tushare接口获取数据
get_hist_data虽然只能获取近三年的数据
但是因为快速不会time_out所以使用它
详细api参考http://tushare.waditu.com/trading.html#id2

```python
df = ts.get_hist_data(stockcode,start,end)
```

展示获取到的数据(dataframe)

```python
df
```

能够看到我设置的时间是从2013-10-19到2016-10-19
然而2013-10-23前的数据没显示出来.另外,在显示的数据中有明显的空缺
比如2016-9-23到2016-9-26之间的数据没有了,这是因为这两天是周末,不管怎么说这种缺失值是要处理的,不然后续工作很难展开

这里使用的是reindex的方法,具体操作如下

```python
df.index
```

看到dataframe的index是index类型,转换成datetimeindex,后面用到,如果不转换的话reindex会出错

```python
df.index = pd.to_datetime(df.index)
```

```python
df.index
```

新建一个有所有时间的index

```python
idx = pd.date_range(start=start, end=end)
```

```python
idx
```

将原有的dataframe通过将原有的dataframe通过新的index复制一份,缺失的用nan表示,赋值给data

```python
data = df.reindex(idx)
```

```python
data
```

看到数据已被成功复制,缺失的地方用nan表示了出来

取出所有的空值对应的行,重新构造为一个dataframe

```python
missing_rows = data.loc[~(data.volume > 0)]
```

loc是获取dataframe相关的行或者列,这里获取的条件是所有data的volume不是大于0的地方
因为交易量不是大于0的地方就是空值,所以成功的取出了所有满足条件的空行

```python
missing_rows
```

```python
missing_rows = missing_rows.loc[:,['volume']]
```

在missing_rows中选定volume项,重新赋给missing_rows

```python
missing_rows
```

将非交易时段的交易量设为0

```python
missing_rows = missing_rows.fillna(0)
```

```python
missing_rows
```

将这个volume的情况在data中更新

```python
data.update(missing_rows)
```

```python
data
```

看到已经成功,接着将所有在产生数据后的所有空缺数据用前一天的数据插值

```python
data.fillna(method='ffill', inplace=True)
```

```python
data
```

可以看到从2013-10-23以后产生了数据,这之后的节日都用之前的数据插值了,比如2013-10-26,2013-10-27两天的数据沿用了2013-10-25的数据,然而交易量变为了0,表明代码成功

接着将一开始空缺的数据进行插值,这部分数据的空缺可能是因为停牌,可能是有其他原因,不管怎么说,将他们设为0

```python
data.fillna(0,inplace=True)
```

```python
data
```

一开始的空缺数据被设为了0

到此,缺失值处理就搞定了,后续就可以将这个data(dataframe)存入数据库,也可以保存为csv
